```python
def configure_plotly_browser_state():
  import IPython
  display(IPython.core.display.HTML('''
        <script src="/static/components/requirejs/require.js"></script>
        <script>
          requirejs.config({
            paths: {
              base: '/static/base',
              plotly: 'https://cdn.plot.ly/plotly-latest.min.js?noext',
            },
          });
        </script>
        '''))
```

```python
configure_plotly_browser_state()

title_str2 = 'Filtered IR ADC over Time for Patient 003-2162 '
fig2 = go.Figure(layout_title_text = title_str2)
fig2.add_trace(go.Scatter(x = Raw.PPG_Datetime[Raw.study_no == '003-2162'][2000:4000], y =  Raw.IR_ADC_bpf[Raw.study_no == '003-2162'][2000:4000], name='Filtered IR ADC'))
fig2.update_xaxes(rangeslider_visible=False)
fig2.update_xaxes(title_text="Date")
fig2.update_yaxes(title_text="IR ADC")
fig2.update_layout(showlegend=True, font=dict(
        size=18
    ))
fig2.show()
```

```python
configure_plotly_browser_state()

title_str2 = 'Raw IR ADC over Time for Patient 003-2162'
fig2 = go.Figure(layout_title_text = title_str2)
fig2.add_trace(go.Scatter(x = Raw.PPG_Datetime[Raw.study_no == '003-2162'][2000:4000], y =  Raw.IR_ADC[Raw.study_no == '003-2162'][2000:4000], name=' Raw IR ADC'))
fig2.update_xaxes(rangeslider_visible=False)
fig2.update_xaxes(title_text="Date")
fig2.update_yaxes(title_text="IR ADC")
fig2.update_layout(showlegend=True, font=dict(
        size=18
    ))
fig2.show()
```
